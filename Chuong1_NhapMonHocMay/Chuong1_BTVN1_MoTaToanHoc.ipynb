**Câu hỏi mở rộng: Theo bạn biểu diễn (hay mô tả) Toán học của đoạn phim (video) là gì?**

---

> Biểu diễn toán học của một đoạn phim (video) là một tensor hạng 4 - **$R^{h*w*c*f}$**. Với :

*   h là chiều cao bức ảnh (height)
*   w là chiều rộng bức ảnh (width)
*   c là tổng số khung hình/ảnh cấu thành video (count)
*   f là số khung hình/ảnh trong mỗi giây (fps)

Đoạn code trích xuất video 

```python
# Tạo đối tượng đọc video
cap = cv2.VideoCapture(os.path.join('HUET.mp4'))

# Vòng lặp - lặp qua từng khung hình để xuất video
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    ret, frame = cap.read()
    cv2.imshow('Frame', frame)
    # Thoát ra khỏi vòng lặp 
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
